```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Concepts_9.csv')
time_a = data['Time']
time_arr = time_a.to_numpy()
concept_a = data['Concept']
concept_arr = concept_a.to_numpy()
pd_a = data['Present_Difficulty']
pd_arr = pd_a.to_numpy()
res_a = data['Response']
res_arr = res_a.to_numpy()
diff_a = pd.Series([''])
```

```python
for i in range(len(concept_arr)):
    if(res_a[i]==0):
        if(pd_a[i]==1):
          diff_a[i] = 1
        else:
          diff_a[i] = pd_a[i]-1  
    else:
      if(concept_arr[i]==1 or concept_arr[i]==4 or concept_arr[i]==7):
        if(pd_arr[i]==1):
         if(time_arr[i]<=10):
            diff_a[i] = 3
         elif(time_arr[i]>10 and time_arr[i]<=25):
             diff_a[i] = 2
         elif(time_arr[i]>25):
            diff_a[i] = 1  
        elif(pd_arr[i]==2):
         if(time_arr[i]<=20):
            diff_a[i] = 3
         elif(time_arr[i]>20 and time_arr[i]<=35):
             diff_a[i] = 2
         elif(time_arr[i]>35):
            diff_a[i] = 1  
        else:
         if(time_arr[i]<=40):
          diff_a[i] = 3
         else:
          diff_a[i] = 2 
        
      elif(concept_arr[i]==2 or concept_arr[i]==5 or concept_arr[i]==8):
       if(pd_arr[i]==1):
        if(time_arr[i]<=15):
            diff_a[i] = 3
        elif(time_arr[i]>15 and time_arr[i]<=30):
             diff_a[i] = 2
        elif(time_arr[i]>30):
            diff_a[i] = 1  
       elif(pd_arr[i]==2):
          if(time_arr[i]<=25):
            diff_a[i] = 3
          elif(time_arr[i]>25 and time_arr[i]<=40):
             diff_a[i] = 2
          elif(time_arr[i]>40):
            diff_a[i] = 1  
       else:
          if(time_arr[i]<=45):
            diff_a[i] = 3
          else:
            diff_a[i] = 2 
      else:
        if(pd_arr[i]==1):
         if(time_arr[i]<=20):
            diff_a[i] = 3
         elif(time_arr[i]>20 and time_arr[i]<=35):
             diff_a[i] = 2
         elif(time_arr[i]>35):
            diff_a[i] = 1  
        elif(pd_arr[i]==2):
         if(time_arr[i]<=30):
            diff_a[i] = 3
         elif(time_arr[i]>30 and time_arr[i]<=45):
             diff_a[i] = 2
         elif(time_arr[i]>45):
            diff_a[i] = 1  
        else:
         if(time_arr[i]<=50):
          diff_a[i] = 3
         else:
          diff_a[i] = 2 
```

```python
data['Difficulty'] = diff_a
```

```python
data.sample(5)
```

```python
data.to_csv('Concepts_9.csv')
```
